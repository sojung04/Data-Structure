###  3.6

[코드 7-13]의 원형 연결 리스트를 이용해 구현한 Deque에서 enqueue()를 구현한 경우, 항상 큐의 맨 뒤에 원소를 삽입하게 된다. 이때 원형 연결 리스트의 특성 상 큐의 맨 뒤에 원소를 삽입하더라도 큐의 사이즈 n과 관계 없이 상수 시간이 걸린다. 따라서 수행 시간 복잡도는 O(1)로 나타낼 수 있다. 마찬가지로 dequeue()을 구현한 경우, 큐의 맨 앞에서 원소를 삭제하므로 큐의 사이즈 n과 관계 없이 상수 시간이 걸린다. 따라서 dequeue()의 수행 시간 복잡도 또한 enqueue()와 마찬가지로 O(1)로 나타낼 수 있다.

###  3.7

enqueue()를 문제 6번에서처럼 원형 연결 리스트로 구현한 것이 아닌 일반적인 연결리스트로 구현하였다면 enqueue의 수행 시간은 큐의 사이즈 n에 비례하여 증가한다. 따라서 수행 시간 복잡도는 O(n)으로 나타낼 수 있다. 반면 dequeue의 경우 문제 6번에서와 마찬가지로 큐의 맨 앞에서 원소를 삭제할 수 있으므로로 여전히 큐의 사이즈 n과 관계 없이 상수 시간이 걸린다. 따라서 dequeue() 수행 시간 복잡도는 O(1)이다.
